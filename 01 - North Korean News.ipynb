```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import re
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'
}

response = requests.get('http://kcna.kp/kcna.user.home.retrieveHomeInfoList.kcmsf;jsessionid=40849AFD078817016A0F2811A558CE3E', headers=headers)
doc = BeautifulSoup(response.text)    
    
```

```python
lines = doc.find_all("div")#(class_='harticle15_subtitle') #(class_="harticle_wrap")
lines
```

```python
# area = doc.find_all(class_="harticle_wrap")
# head = doc.find_all("h4" and "h3")

# for head in area:
#    print(head.text)
```

```python
articles = []
links = doc.find_all(class_="titlebet")
```

```python
for link in links:
    article = {}
    article['headline'] = link.text
    article['link'] = link['onclick']
    article['ID'] = link['onclick'].split('"')[1]
    articles.append(article)
```

```python
df = pd.DataFrame(articles)
```

```python
df.to_csv("nk-news.csv",index=False)
```
