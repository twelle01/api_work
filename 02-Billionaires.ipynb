## 5) What is the most common source/type of wealth? Is it different between males and females?

> **TIP:** You know how to `groupby` and you know how to count how many times a value is in a column. Can you put them together???
> **TIP:** Use percentages for this, it makes it a lot more readable.

```python
round(df.groupby(by="gender").typeofwealth.value_counts() / df.name.count() * 100,1)
#float in %
```

```python
round(df.groupby(by="typeofwealth").count()["year"].sort_values(ascending=False) / df.year.count() *100,1)
#float in %
```

## 7) How much money do these billionaires have in total?

```python
#how to do this organically?

fancy_company = df[(df.company == "Hyatt")\
                   | (df.company == "Oetker-Gruppe")\
                   | (df.company == "S. C. Johnson & Son")\
                   | (df.company == "Walmart")\
                   | (df.company == "Votorantim Group")]
                     
```

```python
fancy_company.groupby(by="company").networthusbillion.sum().sort_values(ascending=False) # on company level
fancy_company.networthusbillion.sum() # total sum
```
