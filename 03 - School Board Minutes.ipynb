````python
import requests
import pandas as pd
import re
from bs4 import BeautifulSoup
````

````python
headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'
}

response = requests.get(' http://www.mineral.k12.nv.us/pages/School_Board_Minutes', headers=headers)
doc = BeautifulSoup(response.text)    
    
````

````python
board_mins = []
minutes = doc.find_all(id="livesite-page-content-left")
````

````python
for minute in minutes:
    rows = minute.find_all("a", href=True)
    for row in rows:
        board_min = {}
        board_min['date'] = row.text
        board_min['url'] = row['href']
        board_mins.append(board_min)
````

````python
df = pd.DataFrame(board_mins)
````

````python
df.tail()
````

````python
data = df[df.date != 'Tweet']
````

````python
data.tail()
````

````python
data['date'] = pd.to_datetime(data.date)
````

````python
data.head()
````

````python
data.to_csv("minutes.csv",index=False)
````
